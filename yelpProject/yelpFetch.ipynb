```python
# Dependencies
import os
import numpy as np
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
from pprint import pprint
from config import yelp_api_key
```

```python
#to get the different categories from categories.json in Resources

restaurant_categories = []

with open('Resources/categories.json') as json_file:
    cats = json.load(json_file)
    
    for category in cats:
        try:
            if category['parents'][0]=='restaurants':
                restaurant_categories.append(category['alias'])
            
        except Exception as e:
            #print(f'Restaurant not found. Skipping...')
            pass  # no operation


print(restaurant_categories)
```

```python
#https://tools.usps.com/zip-code-lookup.htm?bycitystate
cities = {"Carlsbad,CA" : ['92008','92009','92010','92011'], 
          "Encinitas,CA" : ['92024'], 
          "Oceanside,CA" : ['92054','92055','92056','92057','92058'],
          "San Marcos,CA": ['92069','92078']
         }

#change fetch_data to true if you want to fetch the data again
fetch_data = True


#to fetch data once and store in folder
if fetch_data == True:
    limit = 50
    
    #create Resources folder
    if not os.path.exists(f'Resources'):
            os.makedirs(f'Resources')
            
    #items() is an iterator dictionary function         
    for city,zips in cities.items():
        print(f'city: {city}: {len(zips)}')
        zipString = ",".join(zips) # combine zipcodes for a city into a single string separated by commas
        #print(zipString)
        
        if not os.path.exists(f'Resources/{city}'):
            os.makedirs(f'Resources/{city}')

        for category in restaurant_categories:
            totalRecords = limit
            offset = 0 # starting record for the query - useful when more than 50 total records
            done = False
            goodRecords = []
            while done==False:
                try:     
                    #url = f"https://api.yelp.com/v3/businesses/search?term=restaurants&latitude={latitude}&longitude={longitude}&radius=5000&limit=50&categories={category}"
                    url = f"https://api.yelp.com/v3/businesses/search?term=restaurants&location={zipString}&limit={limit}&categories={category}&offset={offset}"
                    r = requests.get(url, headers={"Authorization":"Bearer " + yelp_api_key})
                    recordJson = r.json()
                    totalRecords = recordJson["total"]
                    businesses = recordJson["businesses"]
                    
                    for business in businesses:
                        if business["location"]["zip_code"] in zips:
                            goodRecords.append(business)
                   
                    if len(businesses) < limit: # if less than limit records comes back we are done with this query
                        done = True
                    else:
                        offset = offset + limit; # set offset to get next 'limit' records


                    #with open(f'Resources/{city}/{category}.json', 'w') as outfile:
                    #   json.dump(restaurants_category_json, outfile)
                    #pass

                except Exception as e:
                    print(e)
            
            # done reading all pages of data for this location/category
            jsonOutput = { "businesses": goodRecords, "total": len(goodRecords) }
            print(jsonOutput)
            
            with open(f'Resources/{city}/{category}.json', 'w') as outfile:
                json.dump(jsonOutput, outfile)
                
```
